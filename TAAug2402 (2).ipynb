```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset, load_metric
from transformers import BertTokenizer, RobertaTokenizer, BertForSequenceClassification, RobertaForSequenceClassification, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from sklearn.metrics import accuracy_score, f1_score
import optuna
import torch

# Load the IMDb dataset
dataset = load_dataset('imdb')
train_df = pd.DataFrame(dataset['train'])
test_df = pd.DataFrame(dataset['test'])

# Exploratory Data Analysis (EDA)
def plot_eda(train_df, test_df):
    # Plot the distribution of labels
    plt.figure(figsize=(12, 6))
    sns.countplot(x='label', data=train_df)
    plt.title('Distribution of Labels in Training Set')
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.show()

    # Plot the distribution of review lengths
    train_df['review_length'] = train_df['text'].apply(lambda x: len(x.split()))
    test_df['review_length'] = test_df['text'].apply(lambda x: len(x.split()))

    plt.figure(figsize=(12, 6))
    sns.histplot(train_df['review_length'], bins=50, kde=True, color='blue', label='Train')
    sns.histplot(test_df['review_length'], bins=50, kde=True, color='red', label='Test')
    plt.title('Distribution of Review Lengths')
    plt.xlabel('Number of Words')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

plot_eda(train_df, test_df)

# Tokenize the dataset using the specified tokenizer
def tokenize_dataset(dataset, tokenizer_name):
    if tokenizer_name == 'bert':
        tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    elif tokenizer_name == 'roberta':
        tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
    else:
        raise ValueError(f"Unknown tokenizer name: {tokenizer_name}")

    def tokenize_function(examples):
        return tokenizer(examples['text'], padding='max_length', truncation=True)

    tokenized_datasets = dataset.map(tokenize_function, batched=True)
    tokenized_datasets = tokenized_datasets.remove_columns(["text"])
    tokenized_datasets.set_format("torch")
    return tokenized_datasets

tokenized_datasets_bert = tokenize_dataset(dataset, 'bert')
tokenized_datasets_roberta = tokenize_dataset(dataset, 'roberta')

# Reduce the dataset size for quick experimentation
def reduce_dataset_size(tokenized_datasets, size=500):
    tokenized_datasets['train'] = tokenized_datasets['train'].shuffle(seed=42).select(range(size))
    tokenized_datasets['test'] = tokenized_datasets['test'].shuffle(seed=42).select(range(size))
    return tokenized_datasets

tokenized_datasets_bert = reduce_dataset_size(tokenized_datasets_bert)
tokenized_datasets_roberta = reduce_dataset_size(tokenized_datasets_roberta)

# Initialize the model
def model_init(model_name):
    if model_name == 'bert':
        return BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
    elif model_name == 'roberta':
        return RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=2)

# Compute metrics for evaluation
def compute_metrics(p):
    predictions, labels = p
    preds = predictions.argmax(axis=1)
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='weighted')
    return {'accuracy': acc, 'f1': f1}

# Objective function for hyperparameter tuning
def objective(trial, model_name, tokenized_datasets):
    args = {
        'learning_rate': trial.suggest_float('learning_rate', 1e-5, 5e-5, log=True),
        'per_device_train_batch_size': trial.suggest_categorical('per_device_train_batch_size', [8, 16, 32]),
        'num_train_epochs': trial.suggest_int('num_train_epochs', 1, 3),  # Reduced epochs
        'weight_decay': trial.suggest_float('weight_decay', 1e-2, 0.3, log=True)
    }

    training_args = TrainingArguments(
        output_dir=f'./results/{model_name}',
        evaluation_strategy='epoch',
        learning_rate=args['learning_rate'],
        per_device_train_batch_size=args['per_device_train_batch_size'],
        num_train_epochs=args['num_train_epochs'],
        weight_decay=args['weight_decay'],
        logging_dir='./logs',
    )

    trainer = Trainer(
        model_init=lambda: model_init(model_name),
        args=training_args,
        train_dataset=tokenized_datasets['train'],
        eval_dataset=tokenized_datasets['test'],
        compute_metrics=compute_metrics
    )

    trainer.train()
    eval_metrics = trainer.evaluate()
    return eval_metrics['eval_accuracy']

# Perform hyperparameter search
def perform_hyperparameter_search(model_name, tokenized_datasets):
    study = optuna.create_study(direction='maximize')
    study.optimize(lambda trial: objective(trial, model_name, tokenized_datasets), n_trials=5)  # Reduced number of trials
    return study.best_params

# Hyperparameter tuning for BERT
best_params_bert = perform_hyperparameter_search('bert', tokenized_datasets_bert)
print(f"BERT Best Hyperparameters: {best_params_bert}")

# Hyperparameter tuning for RoBERTa
best_params_roberta = perform_hyperparameter_search('roberta', tokenized_datasets_roberta)
print(f"RoBERTa Best Hyperparameters: {best_params_roberta}")

# Train the model with the best hyperparameters
def train_model(model_name, tokenized_datasets, best_params):
    training_args = TrainingArguments(
        output_dir=f'./results/{model_name}',
        evaluation_strategy='epoch',
        learning_rate=best_params['learning_rate'],
        per_device_train_batch_size=best_params['per_device_train_batch_size'],
        num_train_epochs=best_params['num_train_epochs'],
        weight_decay=best_params['weight_decay'],
        logging_dir='./logs',
    )

    trainer = Trainer(
        model_init=lambda: model_init(model_name),
        args=training_args,
        train_dataset=tokenized_datasets['train'],
        eval_dataset=tokenized_datasets['test'],
        compute_metrics=compute_metrics
    )

    trainer.train()
    return trainer

trainer_bert = train_model('bert', tokenized_datasets_bert, best_params_bert)
trainer_roberta = train_model('roberta', tokenized_datasets_roberta, best_params_roberta)

# Evaluate the model
def evaluate_model(trainer):
    eval_result = trainer.evaluate()
    print(f"Evaluation: Accuracy: {eval_result['eval_accuracy']:.2f}, F1: {eval_result['eval_f1']:.2f}")
    return eval_result

eval_result_bert = evaluate_model(trainer_bert)
eval_result_roberta = evaluate_model(trainer_roberta)

# Save the models
trainer_bert.save_model('./fine_tuned_bert')
BertTokenizer.from_pretrained('bert-base-uncased').save_pretrained('./fine_tuned_bert')

trainer_roberta.save_model('./fine_tuned_roberta')
RobertaTokenizer.from_pretrained('roberta-base').save_pretrained('./fine_tuned_roberta')
```
